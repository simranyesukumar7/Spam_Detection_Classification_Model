Preprocessing of text using NLTK

Text preprocessing using the Natural Language Toolkit (NLTK) involves a series of steps to clean, normalize, and prepare text data for further analysis, such as natural language processing (NLP) tasks. Below are common steps for text preprocessing using NLTK:

1. Tokenization:
Breaking down of text into smaller units called tokens.

2. Stopwords:
Commonly used words in a language that actually do not contribute much meaning to the text.

3. Stemming:
Reducing words by chopping off suffixes or prefixes.

4. Lemmatization:
Similar to stemming, but instead of chopping off parts of words, it reduces words to their base form (lemma) using vocabulary and grammar rules.

5. Lowercasing:
Converting all characters in the text to lowercase.

6. Removing Punctuation:
Eliminating symbols like commas, periods, and question marks from the text.

```python
import nltk
nltk.download('punkt')
nltk.download('punkt_tab')
```

```python
# import nltk
# nltk.download('wordnet')
# nltk.download('omw-1.4')
# nltk.data.path.append("/path/to/nltk_data")
# Initialize stopwords and the stemmer
remove_stopwords = set(stopwords.words('english'))
stemmer = PorterStemmer()

# Function to preprocess and transform text
def transform_text(text):
    text = text.lower()
    #Tokenize the text into words
    tokens = word_tokenize(text)
    #Remove special characters and punctuations from the text
    tokens = [word for word in tokens if word.isalnum()]
    #Remove stopwords
    tokens = [word for word in tokens if word not in remove_stopwords and len(word) > 2]
    #Stemming
    transform_tokens = [stemmer.stem(word) for word in tokens]
    # Join the tokens back into a single string
    return " ".join(transform_tokens)

# Apply the transformation to the 'message' column
df['transformed_text'] = df['Text'].apply(transform_text)
```

# Classification Model 

Label Encoder: 
Converting the categorical values into numerical values

```python
label_encoder = LabelEncoder()
df['Target Variable'] = label_encoder.fit_transform(df['Target Variable'])
```

```python
# Split the data into X and y
X = df['transformed_text']
y = df['Target Variable']

#Splitting the dataset
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)
```

```python
#TF-IDF Vectorizer
tfidf = TfidfVectorizer(max_features=3000)
X_train_tfidf = tfidf.fit_transform(X_train)
X_test_tfidf = tfidf.transform(X_test)
```

```python
#Model initialization
models = {
    "KNN": KNeighborsClassifier(n_neighbors=5),
    "Decision Tree": DecisionTreeClassifier(random_state=42),
    "SVM": LinearSVC(random_state=42),
    "Logistic Regression": LogisticRegression(max_iter=1000, random_state=42),
    "Random Forest": RandomForestClassifier(n_estimators=100, random_state=42),
    "Naive Bayes": MultinomialNB(),
}
```

Model Training and Evaluation
Trains a model and evaluates its performance with classification report and confusion matrix.
1. name (str): Name of the model.
2. model: The model to be trained and evaluated.
3. X_train: Training feature set.
4. y_train: Training target labels.
5. X_test: Testing feature set.
6. y_test: Testing target labels.

```python
def train_and_evaluate_model(name, model, X_train, y_train, X_test, y_test):
    print(f"\nModel: {name}")
    # Train the model
    model.fit(X_train, y_train)
    # Predict on the test set
    y_pred = model.predict(X_test)
    #evaluation metrics
    print("Classification Report:")
    print(classification_report(y_test, y_pred))
    # Plot confusion matrix
    ConfusionMatrixDisplay.from_predictions(y_test, y_pred, cmap='Greens', values_format='d')
    plt.title(f'{name} - Confusion Matrix')
    plt.show()
# Train and evaluate all models
for name, model in models.items():
    train_and_evaluate_model(name, model, X_train_tfidf, y_train, X_test_tfidf, y_test)
```
